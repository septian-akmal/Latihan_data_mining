```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("Data_AHC.xlsx")
print(data)
```

```python
x = np.asarray(data)
```

```python
print(x)
```

```python
Z = linkage(x)
```

```python
print(Z)
```

```python
plt.figure(figsize=(15, 10))
plt.title('Dendrogram Hierarchical Klasterong')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
